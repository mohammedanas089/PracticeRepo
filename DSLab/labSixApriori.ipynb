```python
import pandas as pd
from apyori import apriori
```

```python
df=pd.read_csv("./dataset/groceryNewData.csv",header=None)
```

```python
df.drop([0],axis=1,inplace=True)
df.drop([0],axis=0,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.fillna('0',inplace=True)
```

```python
df.head()
```

```python
baskets=[]
for i in range(len(df)-1):
    each_basket=[]
    for j in range(df.shape[1]-1):
        if df.values[i,j]!='0':
            each_basket.append(df.values[i,j])
    baskets.append(each_basket)
```

```python
baskets
```

```python
analysis=apriori(baskets,min_support=0.003,min_confidence=0.10,min_lift=5,min_length=2)
resultdf=pd.DataFrame(list(analysis))
resultdf
```

```python
X=[]
Y=[]
conf=[]
lift=[]
for i in range(len(resultdf)):
    eachPred=resultdf['ordered_statistics'][i][0]
    print(eachPred)
    X.append(list(eachPred[0]))
    Y.append(list(eachPred[1]))
    conf.append(eachPred[2])
    lift.append(eachPred[3])
```

```python
finaldf=pd.concat([pd.DataFrame(X,columns=["X1"]),pd.DataFrame(Y,columns=["Y1","Y2","Y3","Y4","Y5","Y6","Y7","Y8","Y9","Y10"]),resultdf['support'],pd.DataFrame(conf,columns=["Confidence"]),pd.DataFrame(lift,columns=["Lift"])],axis=1)
#finaldf=pd.concat([pd.DataFrame(X),pd.DataFrame(Y),resultdf['support'],pd.DataFrame(conf),pd.DataFrame(lift)],axis=1)
```

```python
finaldf.fillna('0',inplace=True)
```
