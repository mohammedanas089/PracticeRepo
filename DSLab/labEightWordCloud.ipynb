```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.probability import FreqDist
from wordcloud import WordCloud
```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

```python
stopwordlt=stopwords.words("english")
```

```python
file=open("./dataset/Agatha.txt",'r').read()
```

```python
words=word_tokenize(file)
```

```python
len(words)
```

```python
word_no_punc=[]
for word in words:
    if(word.isalpha()):
        word_no_punc.append(word.lower())
```

```python
clean_words=[]
for word in word_no_punc:
    if word not in stopwordlt:
        clean_words.append(word)
```

```python
len(clean_words)
```

```python
word_string=" ".join(clean_words)
```

```python
wcloud=WordCloud(background_color="white").generate(word_string)
plt.axis("off")
plt.imshow(wcloud)
```

```python
maskimg=np.array(Image.open("./img/R.png"))
plt.axis("off")
wcloud=WordCloud(mask=maskimg,background_color="white", contour_width=1,contour_color='black').generate(word_string)
plt.imshow(wcloud)
```
