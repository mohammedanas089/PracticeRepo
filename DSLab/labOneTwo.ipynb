```python
q1=df['horsepower'].quantile(0.25)
q3=df['horsepower'].quantile(0.55)
iqr=q3-q1
lw=q1-1.5*iqr
hw=q3+1.5*iqr
tmpdf=df[df['horsepower'] < hw]
```

```python
sns.boxplot(y='horsepower',data=tmpdf)
```
